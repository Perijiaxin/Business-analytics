```python
df['VEHICLE_COMBINATION'] = df['VEHICLE TYPE CODE 1'].astype(str) + ' & ' + df['VEHICLE TYPE CODE 2'].astype(str)
vehicle_combination = df['VEHICLE_COMBINATION'].value_counts()

most_common_combination = vehicle_combination.idxmax()
most_common_combination
```

```python
df['NUMBER OF PERSONS KILLED'] = df['NUMBER OF PERSONS KILLED'].fillna(0)
alcohol = df[df['CONTRIBUTING FACTOR VEHICLE 1'] == 'Alcohol Involvement']

# Count the number of fatalities in the alcohol-related crashes
fatalities = alcohol[alcohol['NUMBER OF PERSONS KILLED'] > 0].shape[0]

# Calculate the proportion of these that resulted in a fatality
proportion_fatalities = (fatalities / alcohol.shape[0])
proportion_fatalities
```

```python
evening_rush_hour = df[(df['CRASH DATE_CRASH TIME'].dt.hour >= 16) & (df['CRASH DATE_CRASH TIME'].dt.hour < 19)]

evening_rush_hour_count = evening_rush_hour.shape[0]
all_time_crashes = df.shape[0]

proportion_evening_rush_hour = (evening_rush_hour_count / all_time_crashes)
proportion_evening_rush_hour
```

```python
motorcycle_crashes = df[(df['VEHICLE TYPE CODE 1'].str.contains('MOTORCYCLE', na=False)) |
                        (df['VEHICLE TYPE CODE 2'].str.contains('MOTORCYCLE', na=False))]

motorcycle_crashes.loc[:, 'NUMBER OF PERSONS INJURED'] = motorcycle_crashes['NUMBER OF PERSONS INJURED'].fillna(0)
motorcycle_crashes.loc[:, 'NUMBER OF PERSONS KILLED'] = motorcycle_crashes['NUMBER OF PERSONS KILLED'].fillna(0)

injuries_no_fatalities = motorcycle_crashes[(motorcycle_crashes['NUMBER OF PERSONS INJURED'] > 0) & 
                                            (motorcycle_crashes['NUMBER OF PERSONS KILLED'] == 0)].shape[0]

total_motorcycle_crashes = motorcycle_crashes.shape[0]

proportion = (injuries_no_fatalities / total_motorcycle_crashes) * 100
proportion
```
